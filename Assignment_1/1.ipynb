```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# 1. Generate synthetic data
# We'll create 3 distinct clusters
X, y_true = make_blobs(n_samples=400, centers=3, cluster_std=0.60, random_state=0)

# 2. Initialize and Fit the GMM
# n_components is the number of clusters we expect
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm.fit(X)

# 3. Predict the labels (soft clustering)
labels = gmm.predict(X)
probs = gmm.predict_proba(X) # Returns the probability for each cluster

# 4. Visualize the results
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis', zorder=2)
plt.title("GMM Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Display the calculated means
print("Calculated Means:\n", gmm.means_)
```
